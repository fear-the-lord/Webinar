# Step 1: Importing the Dependencies





```
# Import the necessary dependencies
import cv2
import dlib
from google.colab.patches import cv2_imshow
import imutils
from imutils import face_utils
```

# Step 2: Load the image and the detector

```
# Declare e Path Variable that stores the path of the input image 
PATH = "group-photo.jpg"
```

```
image = cv2.imread(PATH)
```

```
# Load the dlib's frontal face detector model
print("[INFO]Loading Model....")
detector = dlib.get_frontal_face_detector() 
```

# Step 3: Pre-Processing the image

```
# Resize the image before detecting to prevent overflow of the image(sometimes)
frame = imutils.resize(image, width = 500)
# Convert the frame to grayscale (For better edge detection)
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
```

# Step 4: Detect the faces on the image

```
# Detect faces 
rects = detector(gray, 1)
```

# Step 5: Loop over all the faces and draw a bounding box 

```
# Now loop over all the face detections and draw the bounding box
for (i, rect) in enumerate(rects): 
		# Draw a rectangle over the detected face 
		(x, y, w, h) = face_utils.rect_to_bb(rect) 
		cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)	
		# Put a number 
		cv2.putText(frame, "Face#{}".format(i + 1), (x - 10, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
```

# Step 6: Show the Result

```
# Show the output image 
cv2_imshow(frame)
#cv2.imshow("Output", frame )
```
